```python
# Importing Neccessary libraries
import os,re,pathlib
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizerFast
from transformers import DistilBertTokenizer
from transformers import TFDistilBertForSequenceClassification
from transformers import TFDistilBertModel
```

```python
# Loading Pre-trained model and tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=6)

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=5e-5),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# Data file Path
data_path = "data/data.csv"
data_path
```

```python
# Reading the Dataset
df = pd.read_csv(data_path)
df.head()
```

```python
df['label'].value_counts()
# Labels in my dataset
```

```python
# Label encoding applying on y variable.

label_encode = {'sadness':0, 'joy':1, 'fear':2, 'anger':3,'love':4, 'surprise':5}
df['label2'] = df['label'].map(label_encode)
```

```python
# Dividing the dataset into independent and dependent variable.
x = list(df['text'])
y = list(df['label2'])
```

```python
# Filtering the Text Data, ( Text-cleaning )
x_data = []
y_data  = []
wrong_dtypes ={}
for i , j in enumerate(x):
    if type(j)==str:
        x_data.append(x[i])
        y_data.append(y[i])
    else:
        wrong_dtypes[i] = j
```

```python
# Length of dataset.
len(x_data),len(x),len(y_data),len(y)
```

complete prototype of the project, working fine

```python
# Set up your dataset
texts = x_data  # List of text samples
labels = y_data  # List of corresponding sentiment labels (0 or 1)

# Split the dataset into train and test sets
train_size = int(0.8 * len(texts))     # <-- 80% for training
train_texts = texts[:train_size]
train_labels = labels[:train_size]
test_texts = texts[train_size:]
test_labels = labels[train_size:]



# Loading Pre-trained Bert model and corresponding their tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=6)

# Compile the model, to Fine-tuning
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=5e-5),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])



# Tokenize the Training and Testing data.
train_encodings = tokenizer(train_texts, truncation=True, padding=True)
test_encodings = tokenizer(test_texts, truncation=True, padding=True)

# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    train_labels
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    test_labels
))

# Train the model
model.fit(train_dataset.shuffle(1000).batch(16), epochs=2, batch_size=16)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_dataset.batch(16), verbose=2)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

### Saving and Loading Model

```python
# Define path to save your model.
model_path = "Bert-Fine-tuned-model/Trained_model"
model.save(model_path)   # <--- Now we are Saving the Trained your location
```

```python
# Loading the Trained model , which we have saved
model = tf.keras.models.load_model(model_path, custom_objects={'TFDistilBertModel': TFDistilBertModel})
```

### Model Evaluation

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_dataset.batch(16), verbose=2)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
