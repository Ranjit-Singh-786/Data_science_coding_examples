```python
import yfinance as yf
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Load Tesla data
tesla = yf.Ticker("TSLA")

# Fetch historical data for the last year
tesla_data = tesla.history(period="1y")

# Display the first few rows of the data
tesla_data.head()
```

```python
tesla_data.tail()
```

```python
tesla_data.index
```

```python
tesla_data['Open'].plot(figsize=(15,6))
```

```python
## xlimit and y limit
tesla_data['High'].plot(xlim=['2023-11-27','2024-05-10'],figsize=(15,6))
```

```python
tesla_data['High'].plot(xlim=['2023-11-27','2024-05-10'],ylim=[150,300],figsize=(15,6))
```

```python
tesla_data['Open'].plot(xlim=['2023-11-27','2024-05-10'],ylim=[120,280],figsize=(15,6),ls="--",color='g')
```

```python
tesla_data['Open'].plot(xlim=['2023-11-27','2024-05-10'],ylim=[120,280],figsize=(15,6),color='g')
tesla_data['Close'].plot(xlim=['2023-11-27','2024-05-10'],ylim=[120,280],figsize=(15,6),color='k')
tesla_data['High'].plot(xlim=['2023-11-27','2024-05-10'],ylim=[120,280],figsize=(15,6),lw=2,color='r')
plt.legend()
```

```python
tesla_data['Open'].plot(figsize=(15,6),color='g')
tesla_data['Close'].plot(figsize=(15,6),color='k')
plt.legend()
```

```python
tesla_data['Open'].plot(figsize=(15,6),color='g',label='Actual opening')
tesla_data['Open'].rolling(window=5,min_periods=1).mean().plot(figsize=(15,6),color='k',label='SMA')
plt.title("Actual opening with Simple moving average")
plt.legend()
```

### Cumulative average 

```python
tesla_data['Open'].expanding().mean().plot(figsize=(15,6))
plt.title("cumulative average of Open price",fontdict={})
plt.show()
```

```python
tesla_data['Open'].plot(figsize=(15,6),label="Actual Opening price")
tesla_data['Open'].expanding().mean().plot(figsize=(15,6),label="Cumulative Average")
plt.title("cumulative average of Open price",fontdict={})
plt.legend()
plt.show()
```

```python
tesla_data['Open'].plot(xlim=["2024-07-01","2024-12-04"],figsize=(15,6),label="Actual Opening price")
tesla_data['Open'].expanding().mean().plot(xlim=["2024-07-01","2024-12-04"],figsize=(15,6),label="Cumulative Average")
plt.title("cumulative average of Open price",fontdict={})
plt.legend()
plt.show()
```

```python
tesla_data['Close'].plot(xlim=["2024-07-01","2024-12-04"],figsize=(15,6),label="Actual Opening price")
tesla_data['Close'].expanding().mean().plot(xlim=["2024-07-01","2024-12-04"],figsize=(15,6),label="Cumulative Average")
plt.title("cumulative average of Open price",fontdict={})
plt.legend()
plt.show()
```
