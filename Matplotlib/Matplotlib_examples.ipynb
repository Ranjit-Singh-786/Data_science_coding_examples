>> ## Matplotlib Examples

```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
import pandas as pd
import numpy as np
print('libraries imported')
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df2 = sns.load_dataset('iris')
df
# loading the dataset from seaborn to for visualaization using matplotlib
```

```python
df = pd.read_csv('tips.csv')
df.head()
# reading the csv file data
```

### line plot with one axis

```python
plt.plot(df['total_bill'])
plt.title('line plot on total  price')
plt.ylabel('total price')
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(df[df['smoker']=='No']['total_bill'],linewidth=3 ,color='g',marker='*',mfc='g',ms=10 , label='no smoker')
plt.plot(df[df['smoker']=='Yes']['total_bill'],linewidth=3 ,color='r',marker='o',mfc='r',ms=10,label='somker')
plt.legend()
plt.title('Billing comparision in smoker',size=20,color='b')
# plt.axis('off')
plt.show()
```

### line plot with two axis. i sorted both record

```python
plt.plot(df.sort_values('total_bill')['total_bill'],df.sort_values('size')['size'])
plt.title('line plot with two axis')
plt.xlabel('sorted size')
plt.ylabel('sorted price')
plt.show()
```

```python
plt.plot(df['total_bill'],label='total_bill',marker='*',ms=6,mfc='r')
plt.plot(df['size'],label='size')
plt.plot(df['tip'],label='tip')
plt.title('relationship in trend of total bill , size , tip')
plt.legend()
plt.show()
```

### Scatter plot

```python
#first scatter plot
plt.scatter(x= df2['sepal_length'],y= df2['petal_length'])
plt.title('scatter plot')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# second scatter plot
plt.scatter(x= df2['sepal_length'],y= df2['petal_length'],marker='*',color='c')     # 'o' for bullet
plt.title('scatter plot')
plt.xlabel('sepal length')
plt.ylabel('petal_length')
plt.show()
```

```python
plt.scatter(x= df2['sepal_length'],y= df2['petal_length'],marker='*')
plt.title('scatter plot')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# third scatter plot
plt.scatter(df['total_bill'],df['size'],color='r',marker='p')
# plt.axis('off')
plt.title('total bill Vs Size')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12,8))
x = np.random.randint(100, size=(200))
y = np.random.randint(100, size=(200))
colors = np.random.randint(100, size=(200))
sizes = 10 * np.random.randint(100, size=(200))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')          # for cmap=plt.cm.coolwarn

plt.colorbar()     # for show the color bar at right side
plt.title("attractive scatter plot.",fontsize=15,color='r',family='serif')  #this is awesome idea 
plt.show()
```

### Bar plot ()

```python
plt.Figure(figsize=(12,5))
plt.bar(df['sex'],df['total_bill'],color='c')
plt.title('bill comparison gender wise')
plt.show()
```

```python
plt.Figure(figsize=(12,5))
colo = ['r','g']
plt.bar(x=df['sex'],y=df['total_bill'],color=colo)
plt.title('bill comparison gender wise')
plt.show()
```

### Pie plot ()

```python
# Assuming df is your DataFrame
label = list(df['sex'].value_counts().index)
value = list(df['sex'].value_counts().values)

plt.pie(value, labels=label,autopct="%1.2f%%",shadow=True,explode=[0,0.08],wedgeprops={'edgecolor':'k'}) 
plt.title('gender availibility')
plt.show()  # This was missing in your code
```

```python
plt.Figure(figsize=(12,5))
plt.bar(df['sex'],df['total_bill'],color='r',edgecolor='b')
plt.title('bill comparison gender wise')
plt.show()
```

```python
plt.Figure(figsize=(12,5))
plt.bar(df['sex'],df['total_bill'],color='hotpink',edgecolor='c',width = 0.1)
plt.title('Bill comparison gender wise')
plt.show()
# vertical bar plot
```

```python
plt.Figure(figsize=(12,5))
plt.barh(df['sex'],df['total_bill'],color='green',edgecolor='k')
plt.title('Bill comparison gender wise')
plt.show()

# horizontal bar plot
```

### Histogram

```python
plt.figure(figsize=(12,5))
bin = [0,10,20,30,40,50]
plt.hist(df['total_bill'],bins=bin,edgecolor='k',lw=2,label='total bill')
plt.title('Distribution on total bill using histogram.')
# plt.axis('off')
plt.legend()
plt.show()
```
