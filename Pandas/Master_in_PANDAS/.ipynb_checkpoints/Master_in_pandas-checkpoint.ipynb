### DUMPING DATA INTO SQLITE3 DATABASE ON LOCAL

```python
import sqlite3

# Create a cursor
conn = sqlite3.connect('insurance.db')
cursor = conn.cursor()

## CREATING TABLE
table ="""CREATE TABLE insurance_data(age integer, gender varchar(15),bmi float,children integer,smoker varchar(5),region varchar(25),charges float);"""
cursor.execute(table) 

# Define the SQL query with placeholders for INSERT
sql_query = "INSERT INTO insurance_data (age, gender,bmi,children,smoker,region,charges) VALUES (?, ?, ?, ?, ?, ?, ?)"


# Specify the parameter values for the new record
for row in csv_file.iterrows():
    cursor.execute(sql_query, (row[1]['age'],row[1]['sex'],row[1]['bmi'],row[1]['children'],row[1]['smoker'],row[1]['region'],row[1]['charges']))

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

## HANDLING OUTLIERS

```python
## IMPORTED MATPLOTLIB LIBRARY, TO VISUALIZE THE OUTLIERS
import matplotlib.pyplot as plt 
df = pd.read_csv('insurance_data.csv')


df.boxplot(column='charges')
plt.grid('off')
plt.show()
```

### TO REMOVE OUTLIERS
Inter Quartile Range<br>
Find the first quartile, Q1.<br>
Find the third quartile, Q3.<br>
Calculate the IQR. IQR = Q3-Q1.<br>
Define the normal data range with lower limit as Q1–1.5*IQR and upper limit as Q3+1.5*IQR.<br>
Any data point outside this range is considered as outlier and should be removed for further analysis.

```python
q1 , q3  = df['charges'].quantile([0.25,0.75])
q1  ,q3 
```

```python
IQR  = q3 - q1
lower_fence = q1 - (IQR*1.5)
upper_fence = q3 + (IQR * 1.5)
print("lower limit :",lower_fence ,"upper limit :", uper_fence)
```

```python
df.loc[(df['charges'] >= lower_fence) & (df['charges']<=upper_fence)].boxplot(column='charges')
plt.title('Outliers free data by charges ')
plt.show()
```

```python
df.loc[(df['charges'] >= lower_fence) & (df['charges']<=upper_fence)].hist(column='charges')
plt.show()
```

## ANALYSIS on "BMI" COLUMNS, GENDER & SMOKING WISE
BMI is less than 18.5  -->  underweight range <br>
18.5 to 24.9           -->  Healthy Weight range<br>
25.0 to 29.9           -->  overweight range

```python

bmi_analysis = pd.DataFrame(group2['bmi'].agg(minimum_bmi="min",maximum_bmi="max",average_bmi="mean"))
bmi_analysis
```

## MANIPULATING THE RESULT ANLYSIS BY CATEGORY

```python
## MANIPULATED THE RESULT OF THE ANLYSIS BY CLASSES

test = bmi_analysis.reset_index()
bins = [0, 18.5, 24.9, float('inf')]  # Define the bins for underweight, normal, and overweight
labels = ['Underweight', 'Normal', 'Overweight']  # Define the labels for each category

test[['average_bmi', 'minimum_bmi', 'maximum_bmi']] = test[['average_bmi', 'minimum_bmi', 'maximum_bmi']].apply(
    lambda x: pd.cut(x, bins=bins, labels=labels, right=False)
)
test
```
