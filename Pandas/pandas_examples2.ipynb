### rename , astype , groupby

```python
import pandas as pd
```

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.dtypes
```

```python
df['price'].astype('int')
# # changing data type of particular column
```

```python
labels = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fourth Owner Or More':4}
# label encoding using map function
```

```python
df['owner'].map(labels)
```

```python
df.columns
# printing column names
```

```python
df.columns = ['bike_name', 'price', 'city', 'kms_driven', 'owner', 'age', 'bike_power','brand']
# rename column name
```

```python
df.rename({'city':'location'},axis=1)
# rename column name using rename function
```

```python
df.select_dtypes(include='O')
# selecting column based on the data type
# we are selecting only object type of columns
```

```python
df.select_dtypes(exclude='O')
# now we are selecting only numerical data type columns
```

### GROUP BY OPERATION

```python
import pandas as pd
df = pd.read_csv('Used_Bikes.csv')
df.head()
# reading the dataset
```

```python
group = df.groupby('owner')
# grouping data based on the owner
```

```python
group['price'].mean()
# average price of each owner bikes
```

```python
group[['price']].mean()
# printing same info as a dataframe
```

```python
group['kms_driven'].min()
```

```python
df
```

```python
group[['kms_driven','age']].min()
# grouping operation on multiple columns
```

```python
group[['kms_driven','age','price']].mean()
# grouping operation on multiple columns
```

```python
group.agg({'price':'mean','kms_driven':'mean'})
# multiple operation on different columns
```
