```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print('library imported !')
```

```python
#loadint the dataset for visualaization
df = sns.load_dataset('tips')
df.head()
```

### First we will plot all plot with default functionality, and after that we will do some advance customization on plot.

```python
# countplot
sns.countplot(df['sex'])
plt.title('COUNTPLOT')
plt.ylabel('Gender')
plt.xlabel('Counting')
plt.show()
```

```python
sns.lineplot(df['total_bill'])
plt.title('LINE PLOT')
plt.show()
```

```python
sns.scatterplot(x = df['total_bill'],y=df['tip'])
plt.title('SCATTERPLOT')
plt.show()
```

```python
sns.barplot(x=df['time'],y=df['total_bill'])
plt.title('BAR PLOT')
plt.xlabel('TIME')
plt.ylabel('TOTAL BILL')
plt.show()
```

```python
sns.histplot(df['total_bill'])
plt.title('HISTOGRAM')
plt.ylabel('COUNTING')
plt.xlabel('Bill in Bins')
plt.show()
```

```python
sns.displot(df['total_bill'],kind='kde',fill=True) # by default gives histogram
plt.title('Distribution plot')
plt.xlabel('Distribution of total bill.')
plt.show()
```

```python
sns.kdeplot(df['tip'],fill=True)
plt.title('KDE PLOT')
plt.xlabel('Distribution of tips')
plt.show()
```

```python
# distribution with axis line to check the mean , median , mode
sns.set(rc={'figure.figsize':(10,8)})
sns.displot(df['total_bill'],kind='kde',fill=True)
plt.title('Distribution with axis line',fontdict={'fontsize':20})
plt.axvline(df['total_bill'].mean(),color='#d90429',linewidth=3)
plt.axvline(df['total_bill'].median(),color='#d90400',linewidth=3)
plt.axvline(df['total_bill'].std(),color='#85EB1A',linewidth=3)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull())
plt.title('Heatmap to check missing value.')
plt.show()
# there is no missing value.
```

### correlation using heatmap

```python
metrix = df[['total_bill','tip','size']].corr()
metrix
# correlation value calculated using corr() function.
```

```python
sns.set(rc={'figure.figsize':(12,5)})
sns.heatmap(metrix,annot=True)
plt.title('Correlation using heatmap !',color='b',size=20)
plt.show()
```

```python
# assume if total_bill is our Y variable
# and other all our X variables
threshold = 0.72
columns = metrix[metrix['total_bill'] <= threshold].index
print(f"These are all columns are good to training for our model :- {list(columns)}")
## 
```
